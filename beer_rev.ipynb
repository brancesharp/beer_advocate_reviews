# Beer Reviews EDA

This data contains over 1.5 million beer reviews from this Beer Advocate. This dataset can be downlaoded from [Kaggle](https://www.kaggle.com/datasets/thedevastator/1-5-million-beer-reviews-from-beer-advocate). 

Data features:

- brewery_name: The name of the brewery that made the beer. (String)
- review_time: The date and time of the review. (String)
- review_overall: The reviewer's overall rating of the beer on a scale of 1 to 5. (Float)
- review_aroma: The reviewer's rating of the beer's aroma on a scale of 1 to 5. (Float)
- review_appearance: The reviewer's rating of the beer's appearance on a scale of 1 to 5. (Float)
- review_profilename: The reviewer's username. (String)
- beer_style: The style of beer. (String)
- review_palate: The reviewer's rating of the beer's palate on a scale of 1 to 5. (Float)
- review_taste: The reviewer's rating of the beer's taste on a scale of 1 to 5. (Float)
- beer_name: The name of the beer. (String)
- beer_abv: The alcohol by volume of the beer. (Float)

Several of the beers reviewed fall into two beer style categories, so we will seperate these out into two beer style columns 

```python
df[['beer_style1', 'beer_style2']] = df['beer_style'].str.split('/', expand=True)
df
```

### Visualizing the Data

```python
sns.histplot(df['review_overall'])
```

```python
df_sorted_1 = df.sort_values('beer_style1')
plt.figure(figsize=(12,80))
sns.barplot(x='review_overall', y='beer_style1', data=df_sorted_1)
plt.title('beer_style1 / review_overall', fontsize=12)
plt.show()
```

```python
df_sorted_2 = df.sort_values('beer_style2')
plt.figure(figsize=(12,15))
sns.barplot(x='review_overall', y='beer_style2', data=df_sorted_2)
plt.title('beer_style2 / review_overall', fontsize=12)
plt.show()
```
